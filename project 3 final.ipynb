```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
zomato_data=pd.read_csv("zomato.csv")
zomato_data
```

```python
del zomato_data["address"]
del zomato_data["url"]
del zomato_data["menu_item"]
del zomato_data['phone']
```

```python
zomato_data
```

```python
zomato_data.isnull()
```

```python
plt.figure(figsize=(20,8))
sns.pairplot(zomato_data)
```

```python
plt.figure(figsize=(20,8))
sns.set_style(style='whitegrid')
sns.scatterplot(x='name',y='votes',data=zomato_data)
```

# online count

```python
online_offline_count=zomato_data[["online_order"]]
online_offline_count
```

```python
online_offline_count.groupby('online_order').value_counts()
```

```python
plt.figure(figsize=(5,8))
sns.countplot(data=zomato_data,x="online_order")

plt.xticks(rotation=70)
plt.tight_layout()
plt.show()
```

```python
table_booking = zomato_data[["book_table"]]
table_booking
```
