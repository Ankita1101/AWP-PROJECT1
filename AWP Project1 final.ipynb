# yearwise

```python
covid_data['Date'] = pd.to_datetime(covid_data['Date'])
```

```python
covid_data['Year']= covid_data['Date'].dt.year
covid_data['Month']= covid_data['Date'].dt.month
covid_data['Day']= covid_data['Date'].dt.day
```

```python
data_2020 = covid_data[covid_data['Year'] == 2020]
data_2020
```

```python
data_2020.sum()
```

```python
sum_2020_confirmed = data_2020.Confirmed.sum()
sum_2020_confirmed
```

```python
data_2021 = covid_data[covid_data['Year'] == 2021]
data_2021
```

```python
data_2021.sum()
```

```python
sum_2021_confirmed = data_2021.Confirmed.sum()
sum_2021_confirmed
```

# 2020 and 2021 differnce & positivity rate

```python
differnce_con_rate = (sum_2021_confirmed)-(sum_2020_confirmed)
differnce_con_rate
```

```python
positivity_rate = differnce_con_rate/sum_2020_confirmed*100
positivity_rate
```

# Death differnece & death rate

```python
sum_2020_Deaths = data_2020.Deaths.sum()
sum_2020_Deaths
```

```python
sum_2021_Deaths = data_2021.Deaths.sum()
sum_2021_Deaths
```

```python
Differnce_death_rate = sum_2021_Deaths-sum_2020_Deaths
Differnce_death_rate
```

```python
Death_rate=Differnce_death_rate/sum_2020_Deaths*100
Death_rate
```

# Cured Percentage

```python
sum_2020_Cured = data_2020.Cured.sum()
sum_2020_Cured
```

```python
sum_2021_Cured = data_2021.Cured.sum()
sum_2021_Cured
```

```python
Differnce_Cured_rate = sum_2021_Cured-sum_2020_Cured
Differnce_Cured_rate
```

```python
Cured_rate=Differnce_Cured_rate/sum_2020_Cured*100
Cured_rate
```
